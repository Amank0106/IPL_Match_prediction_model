<h1> ML MODEL</h1>

```python
data[:10]
```

```python

data=data.dropna(subset=['city','winner'])
y=data['winner']
x=data.drop(columns=['winner','venue'])
data
```

```python
categorical=['team1','team2','toss_winner','toss_decision','city']
numeric=['season']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.compose  import ColumnTransformer  
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2
                                              )
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical),
        ('num', 'passthrough', numeric)
    ]
)
```

```python
models={
    'LogisticRegression' : LogisticRegression(max_iter=10000),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'DecisionTreeClassifier' : DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(random_state=42)}
```

```python
results=[]
```

```python
for name,model in models.items() :
    pipe=Pipeline([('preprocessor',preprocessor),
              ('model',model)])
    pipe.fit(x_train,y_train)
    pred=pipe.predict(x_test)
    acc=accuracy_score(y_test,pred)
    results.append([name,acc])
    
    
    
    
    
```

```python
x.isnull().sum()
```

```python
modeldata=pd.DataFrame(results,columns=['model','accuracy'])
modeldata
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(data=modeldata,x='model',y='accuracy')
plt.xticks(rotation=45)
plt.savefig('modeldata.png')
```

```python
# pipeline of best performing model
pipe2=Pipeline([('preprocessor',preprocessor),('model',RandomForestClassifier(random_state=42))])
pipe2.fit(x_train,y_train)
pred2=pipe2.predict(x_test)
print(accuracy_score(y_test,pred2))
```

<h3> MAKE CHANGES IN THIS TO PREDICT DIFFERNT GAME SCENERIOS</h3>

```python
new_match = pd.DataFrame({
    'season': [2010],
    'city': ['Chennai'],
    'team1': ['Mumbai Indians'],
    'team2': ['Chennai Super Kings'],
    'toss_winner': ['Chennai Super Kings'],
    'toss_decision': ['bat']
    
})
```

```python

classes = list(pipe2.classes_)
prob = pipe2.predict_proba(new_match)[0]

team1_name = new_match['team1'].iloc[0]
team2_name = new_match['team2'].iloc[0]
p1 = prob[classes.index(team1_name)]
p2 = prob[classes.index(team2_name)]

# Renormalization 
total = p1 + p2
team1_pct = (p1 / total) * 100 if total > 0 else 50.0
team2_pct = (p2 / total) * 100 if total > 0 else 50.0
```

<h3> PREDICTION</h3>

```python

result = team1_name if p1 >= p2 else team2_name

print(f" {team1_name}: {team1_pct:.1f}%  |  {team2_name}: {team2_pct:.1f}%")
print(f" Predicted winner: {result}")
```

<h3> SAVE THE PIPELINE</h3>




```python
import joblib
from datetime import datetime, timezone

bundle = {
    "pipeline": pipe2,
    "accuracy": accuracy_score(y_test, pred2),
    "model_name": "Random Forest Classifier",
    "feature_order": ["season", "city", "team1", "team2", "toss_winner", "toss_decision"],
    "categorical_features": categorical,
    "numeric_features": numeric,
    "classes": sorted(pipe2.classes_.tolist()),
    "n_matches_used": int(len(x)),
    "n_train": int(len(x_train)),
    "n_test": int(len(x_test)),
    "trained_at": datetime.now(timezone.utc).isoformat(),
    "random_state": 42,
}
joblib.dump(bundle, "pipeline.pkl")
print(f"Saved pipeline bundle -> pipeline.pkl (test accuracy: {bundle['accuracy']*100:.2f}%)")
```
